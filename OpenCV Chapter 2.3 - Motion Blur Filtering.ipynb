# Motion Blur Filtering

This exercise uses a convolution kernel to give a motion blurring effect. The kernel takes the neighboring pixels in a certain direction and averages them to get the pixel value for the new image. For example, this 5 x 5 kernel will blur an image in the horizontal direction:

```python
# Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Generate a horizontal motion blur kernel
kernel = np.zeros((5, 5))
kernel[int((5 - 1) / 2), :] = np.ones(5)
kernel = kernel / 5
kernel
```

Here I'm trying a simple horizontal and vertical motion blurring on an image of a bird.

```python
# Read sample image
src = cv2.imread('Images/goose.jpg', cv2.IMREAD_COLOR)

# Kernel size
size = 120

# Generate the horizontal motion blur kernel
horizontal = np.zeros((size, size))
horizontal[int((size - 1) / 2), :] = np.ones(size)
horizontal = horizontal / size

# Generate the vertical motion blur kernel
vertical = np.zeros((size, size))
vertical[:, int((size - 1) / 2)] = np.ones(size)
vertical = vertical / size

# Apply the kernel to the input image
hzOutput = cv2.filter2D(src, -1, horizontal)
vtOutput = cv2.filter2D(src, -1, vertical)

# Display images
src = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)
hzOutput = cv2.cvtColor(hzOutput, cv2.COLOR_BGR2RGB)
vtOutput = cv2.cvtColor(vtOutput, cv2.COLOR_BGR2RGB)

fig = plt.figure(figsize = (17, 10))
ax1 = fig.add_subplot(131)
ax1.set_xticks([])
ax1.set_yticks([])
ax1.set_title('Original Goose')
_ = ax1.imshow(src)
ax2 = fig.add_subplot(132)
ax2.set_xticks([])
ax2.set_yticks([])
ax2.set_title('Horizontal Goose')
_ = ax2.imshow(hzOutput)
ax2 = fig.add_subplot(133)
ax2.set_xticks([])
ax2.set_yticks([])
ax2.set_title('Vertical Goose')
_ = ax2.imshow(vtOutput)
```
