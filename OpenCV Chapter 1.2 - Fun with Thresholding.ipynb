# Fun with Thresholding
Thresholding converts a grayscale image to a binary (two-color) image based on a selected threshold for pixel intensity. It is used to make an image simpler and easier to process.

In the photo below, a threshold of 90 is selected. Any pixel with a value:  

- less than 90 is converted to a pixel value of 0 (black)
- greater than 90 is converted to a pixel value of 1 (white) 

```python
# import libraries
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

### Thresholding photos

```python
# Read an image in grayscale
src = cv2.imread("Images/jill.jpg", cv2.IMREAD_GRAYSCALE)

# Set threshold and maximum value
thresh = 90
maxValue = 255

# Binary threshold
th, dst = cv2.threshold(src, thresh, maxValue, cv2.THRESH_BINARY)

# Display images
fig = plt.figure(figsize = (15, 15))
ax1 = fig.add_subplot(211)
ax1.set_xticks([])
ax1.set_yticks([])
ax1.set_title('Gray Jill')
ax1.imshow(src, cmap = 'gray')
ax2 = fig.add_subplot(212)
ax2.set_xticks([])
ax2.set_yticks([])
ax2.set_title('Thresholded Jill')
ax2.imshow(dst, cmap = 'gray')
```

### Image Segmentation

```python
# Read an image in grayscale
src = cv2.imread("Images/ducks.jpg", cv2.IMREAD_GRAYSCALE)

# Set threshold and maximum value
thresh = 40
maxValue = 255

# Binary threshold (inverse)
th, dst = cv2.threshold(src, thresh, maxValue, cv2.THRESH_BINARY_INV)

# Display images
fig = plt.figure(figsize = (15, 15))
ax1 = fig.add_subplot(211)
ax1.set_xticks([])
ax1.set_yticks([])
ax1.set_title('Grayscale Ducks')
ax1.imshow(src, cmap = 'gray')
ax2 = fig.add_subplot(212)
ax2.set_xticks([])
ax2.set_yticks([])
ax2.set_title('Thresholded Ducks')
ax2.imshow(dst, cmap = 'gray')
```
