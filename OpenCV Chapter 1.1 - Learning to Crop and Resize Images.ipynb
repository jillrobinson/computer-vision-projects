# Learning to Crop and Resize Images

Cropping and resizing are useful for improving processing speed by removing unnecessary pixels.

```python
# Import libraries
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

### Original Image

```python
# Read an image
src = cv2.imread("Images/summer.jpg", cv2.IMREAD_COLOR)
img = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)

print "This photo is " + str(src.shape[1]) + " pixels wide and " + str(src.shape[0]) + " pixels high"

# Display images
fig = plt.figure(figsize = (8, 8))
ax1 = fig.add_subplot(111)
ax1.set_xticks([])
ax1.set_yticks([])
ax1.set_title('Tiny Summer')
_ = ax1.imshow(img)
```

### Cropped Image

```python
# Crop photo
cropped = img[100:1200, 300:1500]

print "This cropped photo is " + str(cropped.shape[1]) + " pixels wide and",
print str(cropped.shape[0]) + " pixels high"

# Display images
fig = plt.figure(figsize = (8, 8))
ax1 = fig.add_subplot(111)
ax1.set_xticks([])
ax1.set_yticks([])
ax1.set_title('Cropped Summer')
_ = ax1.imshow(cropped)
```

### Resized Image (scaling)

```python
scaleX = 0.2
scaleY = 0.2

# Scale down photo
smaller = cv2.resize(img, None, fx = scaleX, fy = scaleY, interpolation = cv2.INTER_LINEAR)

print "This scaled down photo is " + str(smaller.shape[1]) + " pixels wide and",
print str(smaller.shape[0]) + " pixels high"

# Display images
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.set_xticks([])
ax1.set_yticks([])
ax1.set_title('Scaled Down Summer')
_ = ax1.imshow(smaller)
```

### Resized Image (fixed size)

```python
# Resize 
tiny = cv2.resize(img, (300, 300), interpolation = cv2.INTER_LINEAR)

print "This square photo is " + str(tiny.shape[1]) + " pixels wide and",
print str(tiny.shape[0]) + " pixels high"

# Display images
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.set_xticks([])
ax1.set_yticks([])
ax1.set_title('Tiny Summer')
_ = ax1.imshow(tiny)
```

So it looks like if the resized image has a different aspect ratio to the original image, the new image will be distorted to fit the new aspect ratio (no letterboxing).
